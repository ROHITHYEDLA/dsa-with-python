```python
#FACTORIAL
def factorial(n):
  if n==0:
    return 1
  return n*factorial(n-1)
```

```python
#217 Contains duplicates
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        hashset=set()
        for n in nums:
            if n in hashset:
                return True
            hashset.add(n)
        return False
```

```python
#shift 2d grid

class Solution:
    def shiftGrid(self, grid, k):
        m = len(grid)
        n = len(grid[0])
        total = m * n

        k %= total

        arr = []

        for row in grid:
            arr.extend(row)

        arr = arr[-k:] + arr[:-k]

        result = []

        for i in range(0, total, n):
            result.append(arr[i:i + n])

        return result
```

```python
~#27
class Solution:
    def removeElement(self, nums, val):
        i = 0

        for j in range(len(nums)):
            if nums[j] != val:
                nums[i] = nums[j]
                i += 1

        return i
```

```python
#26
class Solution:
    def removeDuplicates(self, nums):
        if not nums:
            return 0

        i = 0

        for j in range(1, len(nums)):
            if nums[j] != nums[i]:
                i += 1
                nums[i] = nums[j]

        return i + 1
```

```python
#25
class Solution:
    def reverseKGroup(self, head, k):
        dummy = ListNode(0)
        dummy.next = head
        groupPrev = dummy

        while True:
            kth = groupPrev
            for _ in range(k):
                kth = kth.next
                if not kth:
                    return dummy.next

            groupNext = kth.next

            # Reverse group
            prev = groupNext
            curr = groupPrev.next

            while curr != groupNext:
                temp = curr.next
                curr.next = prev
                prev = curr
                curr = temp

            temp = groupPrev.next
            groupPrev.next = kth
            groupPrev = temp
```

```python
#24
class Solution:
    def swapPairs(self, head):
        dummy = ListNode(0)
        dummy.next = head

        prev = dummy

        while prev.next and prev.next.next:
            first = prev.next
            second = first.next

            # Swap
            prev.next = second
            first.next = second.next
            second.next = first

            prev = first

        return dummy.next
```
